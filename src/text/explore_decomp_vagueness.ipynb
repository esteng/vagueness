```python
import pandas as pd

# read in 
dataframe = pd.read_csv("/Users/Elias/Downloads/protoroles_eng_udewt/protoroles_eng_ud1.2_11082016.tsv",
                       sep="\t")

# subset train 
dev_df = dataframe[dataframe["Split"] == "dev"]
print(f"subsetted dev")
# subset applicable 
dev_df = dev_df[dev_df["Applicable"] == "yes"]
print(f"subsetted applicable")
# subset grammatical 
dev_df = dev_df[dev_df["Sent.Grammatical"] > 2]
print(f"subsetted grammatical")

```

```python
from decomp import UDSCorpus

top_offenders = dev_df.groupby(["Sentence.ID", "Pred.Token", "Arg.Phrase", "Property"])["Response"].std()
top_offenders = top_offenders[top_offenders > 0.0]

top_offenders = top_offenders.sort_values().tail(100)
corpus = UDSCorpus(split="dev")
```

```python
import re

def get_sent(old_name, split="dev"):
    line_num = old_name.split(" ")[1]
    sent_id = f"ewt-{split}-{line_num}"
    return corpus[sent_id].sentence

for id, val in top_offenders.items():
    sent = get_sent(id[0])
    predicate = re.split("\s+", sent)[int(id[1])]
    prop = id[3]
    print(f"sentence: {sent}, ", val)
    print(f"property: {prop}, predicate: {predicate}, argument: {id[2]}")
    print()

```
